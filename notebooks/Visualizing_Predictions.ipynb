# Imports

```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import plotly
import plotly.express as px
import umap
import umap.plot

# Sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import RobustScaler, PowerTransformer

import warnings; warnings.simplefilter('ignore')

from google.colab import output
output.enable_custom_widget_manager()
```

# Environment Definitions and Dataset Loading/Scaling/Transforming

```python
PATH = 'drive/MyDrive/Kaggle/Clustering_072022/'

# data = pd.read_csv(f'{PATH}src/data.csv', index_col='id')
df = pd.read_csv(PATH+'src/data_removed.csv', index_col='id')
preds = pd.read_csv(PATH+'submissions/BGMM_tol0.001_mi200_ni3_optimizedSViter75.csv')
cat_feats = df.columns[df.dtypes=='int'].tolist()
num_feats = df.columns[df.dtypes=='float'].tolist()
df[num_feats] = RobustScaler().fit_transform(df[num_feats])
df_scaled = pd.DataFrame(PowerTransformer().fit_transform(df), columns=df.columns)
```

# PCA 3D Scatter Plots

## Scaled/Transformed Dataset

```python
pca = PCA(n_components=3)
components = pca.fit_transform(df_scaled)
pca_results = pd.DataFrame(components, columns=[f'PCA{i+1}' for i in range(3)])

# 3D scatterplot
fig = px.scatter_3d(
    pca_results, x='PCA1', y='PCA2', z='PCA3', opacity = 1,
    title='PCA plot in 3D',
    width=800, height=600
)
fig.show()
```

```python
preds = pd.read_csv(PATH+'submissions/BGMM_tol0.001_mi200_ni3_optimizedSViter75.csv')

pca = PCA(n_components=3)
components = pca.fit_transform(df_scaled)
pca_results = pd.DataFrame(components, columns=[f'PCA{i+1}' for i in range(3)])
pca_results['labels'] = preds['Predicted']

# 3D scatterplot
fig = px.scatter_3d(
    pca_results, x='PCA1', y='PCA2', z='PCA3',  size=0.01 * np.ones(pca_results.shape[0]), opacity=0.2,
    color = 'labels',
    title='PCA plot in 3D',
    width=800, height=600
)
fig.show()
```

# UMAP

```python
%%time
um = umap.UMAP(n_components=3)
components_umap = um.fit_transform(df_scaled)

fig = px.scatter_3d(
    components_umap, x=0, y=1, z=2, color=preds['Predicted'], opacity = 0.9,
    title='UMAP plot in 3D',
    labels={'0': 'comp. 1', '1': 'comp. 2', '2': 'comp. 3'},
    width=800, height=600
)
fig.show()
```
